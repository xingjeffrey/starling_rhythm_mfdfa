## Parse t-statistics

```python
import scipy.stats as stats
import numpy as np
import statsmodels.stats.multitest
```

```python
ae_df['MF_range'].values[0][0]
```

```python
ae_df['h_expos'][0][0]
```

```python
total_ts = []
total_p = []
total_sig = []
total_bonf_p = []

for mf_range, iaaft_range in zip(ae_df['MF_range'].values, ae_df['mf_iaaft_range']):
    ts = []
    p = []
    for MF_range, mf_iaaft_range in zip(mf_range, iaaft_range):
        test = stats.ttest_1samp(mf_iaaft_range, MF_range)
        ts.append(test[0])
        p.append(test[1])
        
    bonferroni = statsmodels.stats.multitest.multipletests(p, alpha = 0.05, method = 'bonferroni')

    total_ts.append(ts)
    total_p.append(p)
    total_sig.append(bonferroni[0])
    total_bonf_p.append(bonferroni[1])
    
ae_df['ts'] = total_ts
ae_df['p'] = total_p
ae_df['sig'] = total_sig
ae_df['bonf_p'] = total_bonf_p
```
